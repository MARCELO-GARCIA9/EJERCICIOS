```python
import pandas as pd
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/bwt.csv')
```

```python
df
```

```python
df['peso_guagua_kg']=df['bwt']/35.274
df['altura_mama_metro']=df['height']/39.37
df['peso_mama_kg']=df['weight']*0.453592
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
df[df.parity==1]['gestation'].hist(cumulative=True,density=1, bins=100,histtype='step',label='primogenito')
df[df.parity==0]['gestation'].hist(cumulative=True,density=1, bins=100,histtype='step',label='otro')
plt.title('Periodo de gestacion')
plt.xlabel('Dias')
plt.legend(loc="upper left")
plt.show()
```

```python
df['altura_mama_metro'].hist(cumulative=True,density=1, bins=100 )
```

```python
muestra=df.sample(frac=0.3, replace=True)
muestra['peso_guagua_kg'].hist(cumulative=True,density=1, bins=100 )
```

Se puede apreciar que la funcion de distribucion acumulada para el peso de los bebes se distribuye normal ya que el grafico es como una S bastante sutil. Tambien se puede estimar que la media se encuentra
